```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
path=("bridge.jpg")
img=cv2.imread(path)
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
vf=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
```

```python
def convolution(img,f):
    convolved_image=np.zeros((img.shape[0]-f.shape[0]+1,img.shape[1]-f.shape[1]+1))
    for rp in range(convolved_image.shape[0]):
        for cp in range(convolved_image.shape[1]):
            sliced_image=img[rp:rp+f.shape[0],cp:cp+f.shape[1]]
            convolved_image[rp][cp]=np.sum(sliced_image*f)
    return convolved_image
```

```python
plt.imshow(convolution(img,vf),cmap='gray')
```
